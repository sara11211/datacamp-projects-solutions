**1- What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.**

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("datasets/netflix_data.csv")
```

```python
import numpy as np

# filter data to store '1990s movies'
netflix_90s_movies = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999) & (netflix_df['type'] == 'Movie')]

#visualize the duration distribution using a histogram
netflix_90s_movies['duration'].hist()
plt.title('1990s movies duration distribution')
plt.xlabel('duration (min)')
plt.ylabel('number of movies')
plt.show()
```
